```python
# rename AI_release_date to test_release_date
df.rename(columns={'AI_release_date': 'test_release_date'}, inplace=True)

df
```

```python
# from the test_release_date if there is a value then replace that value to the AI_Release_Date. Ignore the null values from the test_release_date.
df.loc[df['test_release_date'].notnull(), 'AI_Release_Date'] = df['test_release_date']
```
